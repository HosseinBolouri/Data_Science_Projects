# Handwritten Digit Classification (Neural Network from Scratch + TensorFlow)
Multi-class (0–9) digit recognition using a fully-connected neural network. Demonstrates:
- Manual softmax verification vs TensorFlow
- Loading preprocessed 20x20 grayscale digits (flattened to 400 features) from X.npy / y.npy
- Model architecture: Dense(25, relu) -> Dense(15, relu) -> Dense(10, linear logits)
- SparseCategoricalCrossentropy (from_logits=True) loss, Adam optimizer
- Unit test of model structure and forward pass
- Weight inspection
- Training for 100 epochs and plotting loss
- Single example prediction (digit “2”) with probabilities
- Grid of random predictions (true label vs predicted)
- Error counting (accuracy summary)

```python
# --------------------------------------------------
# Imports / Setup
# --------------------------------------------------
# Multi class classification problem (hand written digit recognition from 0 to 9)
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import linear, relu, sigmoid
import matplotlib.pyplot as plt
import logging
logging.getLogger("tensorflow").setLevel(logging.ERROR)
tf.autograph.set_verbosity(0)
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
from IPython.display import display

# --------------------------------------------------
# Softmax (manual) and quick test
# --------------------------------------------------
def my_softmax(z):
    """ Softmax converts a vector of values to a probability distribution.
    Args:
      z (ndarray (N,))  : input data, N features
    Returns:
      a (ndarray (N,))  : softmax of z
    """
    ez = np.exp(z)
    a = ez/np.sum(ez)
    return a

z = np.array([1., 2., 3., 4.])
a = my_softmax(z)
atf = tf.nn.softmax(z)
print(f"my_softmax(z):         {a}")
print(f"tensorflow softmax(z): {atf}")

plt.close("all")

# --------------------------------------------------
# Data loading
# --------------------------------------------------
def load_data():
    X = np.load("X.npy")
    y = np.load("y.npy")
    return X, y

# load dataset
X, y = load_data()

# --------------------------------------------------
# Dataset inspection
# --------------------------------------------------
print('The shape of X is: ' + str(X.shape))
print('The shape of y is: ' + str(y.shape))
print('The first element of X is: ', X[0])
print('The first element of y is: ', y[0,0])
print('The last element of y is: ', y[-1,0])

# --------------------------------------------------
# Sample grid of images
# --------------------------------------------------
m, n = X.shape
fig, axes = plt.subplots(8,8, figsize=(5,5))
fig.tight_layout(pad=0.13, rect=[0, 0.03, 1, 0.91])
fig.suptitle("Label, image", fontsize=14)
display(fig)
for ax in axes.flat:
    random_index = np.random.randint(m)
    X_random_reshaped = X[random_index].reshape((20,20)).T
    ax.imshow(X_random_reshaped, cmap='gray')
    ax.set_title(y[random_index,0])
    ax.set_axis_off()
plt.show()

# --------------------------------------------------
# Model definition
# --------------------------------------------------
tf.random.set_seed(1234)
model = Sequential(
    [
        tf.keras.layers.InputLayer((400,)),
        tf.keras.layers.Dense(25, activation="relu", name="L1"),
        tf.keras.layers.Dense(15, activation="relu", name="L2"),
        tf.keras.layers.Dense(10, activation="linear", name="L3")
    ],
    name="my_model"
)
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True))
model.summary()

# --------------------------------------------------
# Unit test function
# --------------------------------------------------
def test_model(model, num_classes, input_dim):
    """
    Basic structural/unit test for the model.
    Checks input/output shapes and runs a forward pass.
    """
    assert model.input_shape == (None, input_dim), f"Input shape mismatch: {model.input_shape} != (None, {input_dim})"
    assert model.output_shape == (None, num_classes), f"Output shape mismatch: {model.output_shape} != (None, {num_classes})"
    x_dummy = np.random.rand(2, input_dim).astype(np.float32)
    y_logits = model(x_dummy)
    assert y_logits.shape == (2, num_classes), "Forward pass output shape incorrect"
    y_probs = tf.nn.softmax(y_logits, axis=1).numpy()
    print("Logits sample:", y_logits.numpy()[0])
    print("Softmax probabilities sample:", y_probs[0])
    print("Sum of probs (should be ~1):", y_probs[0].sum())
    print("Model unit test passed.")

test_model(model, 10, 400)

# --------------------------------------------------
# Examine weights
# --------------------------------------------------
[layer1, layer2, layer3] = model.layers
W1,b1 = layer1.get_weights()
W2,b2 = layer2.get_weights()
W3,b3 = layer3.get_weights()
print(f"W1 shape = {W1.shape}, b1 shape = {b1.shape}")
print(f"W2 shape = {W2.shape}, b2 shape = {b2.shape}")
print(f"W3 shape = {W3.shape}, b3 shape = {b3.shape}")

# --------------------------------------------------
# Compile with optimizer and train
# --------------------------------------------------
model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
)
history = model.fit(
    X, y,
    epochs=100
)

# --------------------------------------------------
# Plot training loss
# --------------------------------------------------
def plot_loss_tf(history):
    """Plot training loss from a Keras History object."""
    if 'loss' not in history.history:
        print("No 'loss' key found in history.history")
        return
    plt.figure(figsize=(5,3))
    plt.plot(history.history['loss'], label='loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training Loss')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_loss_tf(history)

# --------------------------------------------------
# Single prediction example
# --------------------------------------------------
def display_digit(x):
    plt.figure(figsize=(2,2))
    plt.imshow(x.reshape(20,20).T, cmap='gray')
    plt.axis('off')
    plt.show()

image_of_two = X[1015]
display_digit(image_of_two)

prediction = model.predict(image_of_two.reshape(1,400))
print(f" predicting a Two: \n{prediction}")
print(f" Largest Prediction index: {np.argmax(prediction)}")

prediction_p = tf.nn.softmax(prediction)
print(f" predicting a Two. Probability vector: \n{prediction_p}")
print(f"Total of predictions: {np.sum(prediction_p):0.3f}")

yhat = np.argmax(prediction_p)
print(f"np.argmax(prediction_p): {yhat}")

# --------------------------------------------------
# Grid of predictions vs labels
# --------------------------------------------------
m, n = X.shape
fig, axes = plt.subplots(8, 8, figsize=(5,5))
fig.tight_layout(pad=0.13, rect=[0, 0.03, 1, 0.91])
fig.suptitle("Label, yhat", fontsize=14)
display(fig)
for ax in axes.flat:
    random_index = np.random.randint(m)
    X_random_reshaped = X[random_index].reshape((20,20)).T
    ax.imshow(X_random_reshaped, cmap='gray')
    logits = model(X[random_index].reshape(1,400))
    probs = tf.nn.softmax(logits, axis=1).numpy()[0]
    yhat = np.argmax(probs)
    ax.set_title(f"{y[random_index,0]},{yhat}", fontsize=10)
    ax.axis('off')
plt.show()

# --------------------------------------------------
# Error count
# --------------------------------------------------
def display_errors(model, X, y):
    """
    Count misclassified samples.
    y can be shape (m,) or (m,1) with integer class labels.
    """
    y_true = y.ravel()
    logits = model.predict(X, batch_size=128, verbose=0)
    y_pred = np.argmax(logits, axis=1)
    return int(np.sum(y_pred != y_true))

errors = display_errors(model, X, y)
print(f"{errors} errors out of {len(X)} images (accuracy={(len(X)-errors)/len(X):.3f})")
```

## Summary & Guidance
This notebook trained a 3-layer dense network on flattened 20x20 digit images, verified softmax outputs, inspected weights, evaluated predictions, and reported total errors (accuracy). 

To adapt:

- Modify layer sizes or add regularization (dropout) for generalization.
- Add metrics: confusion matrix, per-class precision/recall.
